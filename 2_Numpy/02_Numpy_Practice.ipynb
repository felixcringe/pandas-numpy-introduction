## Part 2 — Manipulating Arrays

### Challenge 5: Array with mixed-type input

Create a NumPy array from the values `(1, 3, '4', 7, 12, '0')`. Force the data type to **integer** using the `dtype` parameter.

> Notice that the string values `'4'` and `'0'` are silently cast to integers. NumPy arrays are homogeneous — all elements must share the same type.

````python
# YOUR CODE HERE
my_arr = np.array((1, 3, '4', 7, 12, '0'), dtype=int)
````

### Challenge 6: Inspect dtype and shape

Using the array you created in Challenge 5, check:
- The **data type** of the elements
- The **shape** of the array

````python
# YOUR CODE HERE
np.shape(my_arr)
my_arr.dtype
````

### Challenge 7: Update a value by index

Update the **4th element** (index 3) of the array from Challenge 5 to the value `30`.

````python
# YOUR CODE HERE
my_arr[3] = 30
````

### Challenge 8: Reshape to a 2×3 matrix

Reshape the array from Challenge 7 into a **2-row, 3-column** matrix. Print the result.

````python
# YOUR CODE HERE
my_arr = my_arr.reshape(2, 3)
print(my_arr)
````

### Challenge 9: Row-wise addition

Using the 2×3 matrix from Challenge 8:
- Add **8** to every value in the **first row**
- Add **12** to every value in the **second row**

> Hint: index the rows and assign back in place.

````python
# YOUR CODE HERE
print(my_arr - [[8], [0]])
print(my_arr + [[0],[12]])
````

### Challenge 10: Column-wise multiplication

Multiply the columns of your matrix:
- **First column** × 2
- **Second column** × 3
- **Third column** × 4

> Hint: you can do this in one line using broadcasting with a 1D array `[2, 3, 4]`.

````python
# YOUR CODE HERE
my_arr[:, 0] * 2

my_arr * np.array([2, 3, 4])
````

### Challenge 11: Row sums

Calculate the **sum of all values in each row** of your matrix. You should get two totals — one per row.

````python
# YOUR CODE HERE
my_arr.sum(axis=1)
````

### Challenge 12: Column maximums

Find the **largest value in each column** of your matrix. You should get three values — one per column.

````python
# YOUR CODE HERE
my_arr.max(axis=0)
````

### Challenge 13: Index a single element

Extract the value from the **second column, first row** of your matrix.

> Reminder: NumPy uses `array[row, column]` indexing.

````python
# YOUR CODE HERE
my_arr[0, 1]
````

### Challenge 14: Find the index of a specific value

Find the **index** of the value `48` in your matrix.

> Hint: use `np.where()`. Note that depending on how your array evolved across previous challenges, your value at a given position may differ — first check what values are currently in the matrix.

````python
# YOUR CODE HERE
print(my_arr)
np.where(my_arr == 12)
````
